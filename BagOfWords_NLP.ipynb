```python
import nltk
from nltk. stem import PorterStemmer
from nltk.corpus import stopwords
```

Tokenization

paragraphs are conerted into sentences then words

```python
nltk.download('punkt_tab')
sentences=nltk.sent_tokenize(paragraph)
```

```python
print(sentences)
```

```python
type(sentences)
```

Stemming

finds the root context of the word.Example:(history,historical)->histori

```python
stemmer=PorterStemmer()
stemmer.stem('historical')
```

Lemmatization
same as stemming ,but the output doesnt lose meaning.Example:(history,historical)->history

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
lemmatizer.lemmatize('history')
```

cleaning the sentences using regular expression to remove special characters

```python
import re
```

```python
corpus=[]
for i in range(len(sentences)):
  text=re.sub('[^a-zA-Z]',' ',sentences[i])
  text=text.lower()
  corpus.append(text)
```

```python
corpus
```

```python
stopwords.words('english')
```

```python
nltk.download('stopwords')
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

```python
#lemmatization
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(lemmatizer.lemmatize(word))
```
